# Interpolate NaN values from mobile EEG channels for analysis
-- the device obtains NaN values for all of the channels when it detects a blink or jaw clench. Will lead to all four channels obtaining Nans at the same time.

-- I do not intend to clean outliers/ dropped signal since those cannot be accounted for in real-time. The NaNs are an exception because they are caused by artifact detection from the EEG.

```python
def NaNsFromChannel(df, data):
    nans = [df.index[i] for i, value in enumerate(data) if np.isnan(value)]
    return nans

RAW_AF7_nans = NaNsFromChannel(data_5_21, data_5_21.RAW_AF7)
RAW_AF8_nans = NaNsFromChannel(data_5_21, data_5_21.RAW_AF8)
RAW_TP9_nans = NaNsFromChannel(data_5_21, data_5_21.RAW_TP9)
RAW_TP10_nans = NaNsFromChannel(data_5_21, data_5_21.RAW_TP10)
```

### Quick clean via interpolation on NaN

-- Interpolate over NaNs

```python
def interpolateNaNs(channelData):
    nans =  np.isnan(channelData)
    x = lambda z: z.to_numpy().nonzero()[0]
    
    interpolated = channelData.copy()
    interpolated[nans] = np.interp(x(nans), x(~nans), channelData[~nans])
    return interpolated

data_5_21['clean_RAW_AF7'] = interpolateNaNs(data_5_21.RAW_AF7)
data_5_21['clean_RAW_AF8'] = interpolateNaNs(data_5_21.RAW_AF8)
data_5_21['clean_RAW_TP9'] = interpolateNaNs(data_5_21.RAW_TP9)
data_5_21['clean_RAW_TP10'] = interpolateNaNs(data_5_21.RAW_TP10)
```

# Analysis

Transform to Time-Frequency Decomposition via STFT. DC current at 0 to 1 Hz excluded. There is a notch filter for 60 Hz intervals. Seems to be a cycle, but still not visible to eye.

The EEG microvoltage sampling rate is 256. We want the app to be able to recieve power data at a rate of 10 Hz (10 data points per seconds). The time step will need to be approximately 25.6, but it will be kept at 25 for ease of use. This make power data have a rate of approx. 10.24 Hz.

```python
def stft(data, time, fs = 256, window = 'hann'):
    f, t, Zxx = signal.stft(data, fs = fs, window = window, nperseg = fs * 5, 
                           noverlap = fs * 5 - 25) 
    # set it to recieve the power data points at a rate of 10 Hz.
    plt.figure(figsize = (15, 10))
    time = pd.to_datetime(np.linspace(time[0].value, time[-1].value, 
                   t.shape[0]))
    plt.pcolormesh(time, f, np.log10(np.abs(Zxx)), cmap = plt.cm.jet)
    plt.ylim([0.5, fs / 2])
    cbar = plt.colorbar()
    cbar.ax.set_title('log(uv ** 2)')
    plt.title('TF Decomposition')
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (hh:mm:ss)')
    
    plt.show()
    return f, time, Zxx

f, t, Zxx = stft(AF7, data_5_21.index)
```

Sleep can be categorized by the type of brain waves, the presence of certain frequencies in the signal. Slow wave sleep (SWS) is often categorized by an increased presence of delta power (1-4Hz), so that was investigated. 

-- as seen in the data (battery restriction only allowed up to certain amount of time unfortunately), there is a cyclic pattern of sleep depth that forms. We can visualize a trend from light sleep to SWS (deep sleep, characterized by increased delta power), back to light sleep, and back to deep sleep until the Muse EEG battery fails. This can be seen in the pwer log-transformed.

The goal is to be able to determine the sleep stage every 5 minutes via the power data. To see how that trend will look like, we will take a rolling mean to represent where the sleep cycle is every 5 minutes. 

rolling mean: 10 power data points per second * 60 seconds * 5 minutes = 3000 data points.

```python
def deltaPowerTimeSeries(f, t, Zxx):
    delta_frequency = np.zeros([t.shape[0]])
    for index, freq in enumerate(f):
        if (freq >= 1 and freq <= 4.0): # Hz 
            delta_frequency += np.abs(Zxx[index, :])
    # filter edge artifacts from start and end time steps!
    power_at_Delta = ((delta_frequency))[25:-25] 
    time = t[25:-25]
    return time, power_at_Delta

time, power_at_Delta = deltaPowerTimeSeries(f,t,Zxx)
```

```python
plt.figure(figsize = (15, 12))
plt.subplot(2, 1, 1)
plt.plot(time, power_at_Delta)
plt.plot(time, pd.Series(power_at_Delta).rolling(3000).mean())
plt.legend(['Raw Delta Band', 'Rolling Mean Delta Band'])
plt.title('Time series for Delta Band Power')
plt.ylabel('uV ** 2')

plt.subplot(2, 1, 2)
plt.plot(time, pd.Series(power_at_Delta).rolling(3000).mean(), 'C1')
plt.title('Rolling Mean of Delta Band power')
plt.xlabel('Time (hh:mm:ss)')
plt.ylabel('uV ** 2')
plt.show()
```

Log-transformed Power

```python
plt.figure(figsize = (15, 12))
plt.subplot(2, 1, 1)
plt.plot(time, np.log10(power_at_Delta))
plt.plot(time, pd.Series(np.log10(power_at_Delta)).rolling(3000).mean())
plt.legend(['Raw Log-Transformed Delta Band', 'Rolling Mean Log-Transformed Delta Band'])
plt.title('Time series for log-transformed Delta Band Power')
plt.ylabel('log10(uV ** 2)')

plt.subplot(2, 1, 2)
plt.plot(time, pd.Series(np.log10(power_at_Delta)).rolling(3000).mean(), 'C1')
plt.title('Rolling Mean Log-Transformed Delta Band')
plt.ylabel('log10(uV ** 2)')
plt.xlabel('Time (hh:mm:ss)')
plt.show()
```

# Proposed Real-Time Functionality

*  Taking the rolling mean was only for the purpose of enhancing our view of the sleep depth changed over time.
* In real-time, we used a threshold to distinguish the sleep depthness. Above the threshold, will be when we will not wake an individual and below the threshold will be when we are allowed to wake an individual. 
* In the app since we will not use a rolling mean, it will be set up where there is a vote for every 5 minutes. If more power points are below the threshold for the majority of 5 minutes, the person will not be disturbed. Otherwise, the alarm will ring. The threshold shall not be relevant for the wake decision until the desired time window is reached. 
* The power data retrieval will also not start until theres enough points to undergo the first fourier transform. No zero padding or boundary extension for beginning while in application. We will also take absolute value of log-transform so it is on a positive scale. 
* Threshold will be average between the mean of top 50 values and bottom 50 lowest values (to bypass some real-time artifacts influencing wake threshold). These values are collected from the data prior to the sleep window. 
* There will be a UI slider on the app to indicate how conservative or liberal the user wants the system to be in making the decision to wake an individual during the alarm window. 
* Leaning towards conservative means making the threshold higher. A decision will ensure a user is awakened during light sleep, but it also carries the risk of being awakened at the last second of the time window instead if the threshold criteria was not reached.
* Leaning towards a liberal system means making the threshold lower, and a highly liberal system will risk possibly being awakened during SWS (Deep Sleep) in the time window.


3000 points rolling mean to visualize and mimic dominant voting of power for five minute period.

We can visualize the change better with rolling mean, but lets see this concept in true data form as well.

```python
method1_r = np.abs(pd.Series(np.log10(power_at_Delta)).rolling(3000).mean())
sort_method1_r = np.sort(method1_r[~np.isnan(method1_r)].copy()
                         , kind='mergesort')
avg_method1_r = np.mean(np.concatenate([sort_method1_r[:50], sort_method1_r[-50:]]))

plt.figure(figsize = (15, 12))
plt.subplot(2, 1, 1)
plt.plot(time, method1_r, 'C1')
plt.hlines(y=avg_method1_r, xmin = time[0], xmax=time[-1])
plt.text(time[0], np.amax(method1_r), 'Light Sleep')
plt.text(time[-1000], np.amax(method1_r), 'Light Sleep')
plt.text(time[int(time.shape[0] / 2)], np.amin(method1_r), 'Deep Sleep')
plt.title('Rolling Mean Log-Transformed Delta Band')
plt.ylabel('log10(uV ** 2)')
plt.xlabel('Time (hh:mm:ss)')


method1 = np.abs(np.log10(power_at_Delta))
sort_method1 = np.sort(method1[~np.isnan(method1)].copy()
                         , kind='mergesort')
avg_method1 = np.mean(np.concatenate([sort_method1[:50], sort_method1[-50:]]))

plt.subplot(2, 1, 2)
plt.plot(time, method1, 'C1')
plt.hlines(y=avg_method1, xmin = time[0], xmax=time[-1],
          linewidth=4)
plt.title('Log-Transformed Delta Band')
plt.ylabel('log10(uV ** 2)')
plt.xlabel('Time (hh:mm:ss)')
plt.show()
```

# Future Directions Ideas:

* Along with threshold, there should be an attempt to predict future values to see if there is a more optimal time to wake up during the sleep window.
    * This will also allow us to make the decision to wake someone now rather than later.
    * How? With a Recurrent Neural Network. We input a certain number of data samples in return of a predicted series of future outputs.
   
* Complications:This will require many sleep trials for a single user. Data may also be really robust for EEG, but it may be fun to try out.

The following is an example of the idea.

Import Deep Learning Library - Keras

```python
from keras.layers import SimpleRNN, Dense
from keras.models import Sequential
import tensorflow as tf
from keras.utils import plot_model
```

Create time series simulation and plotting functions

```python
def simulate_time_series(batch_size, n_steps):
    freq1 , freq2, offsets1, offsets2 = np.random.rand(4, batch_size, 1)
    time = np.linspace(0, 1, n_steps)
    series = 0.5 * np.sin((time - offsets1) * (freq1 * 10 + 10))  #   wave 1
    series += 0.2 * np.sin((time - offsets2) * (freq2 * 20 + 20)) # + wave 2
    series += 0.10 * np.sin((time - offsets2) * (freq2 * 20 + 20)) # + wave 3
    series += 0.5 * (np.random.rand(batch_size, n_steps) - 0.5)   # + noise
    return series[..., np.newaxis].astype(np.float32)

def plot_series(series, y=None, y_pred=None, x_label="Time", y_label="uV",
               num_predictions = 0):
    plt.plot(series, ".-")
    plt.xlabel(x_label, fontsize=16)
    plt.ylabel(y_label, fontsize=16, rotation=0)
    plt.hlines(0, 0, series.shape[0] - 1 + num_predictions, linewidth=1)
    return

def plot_multiple_forecasts(X, Y, Y_pred):
    n_steps = X.shape[1]
    ahead = Y.shape[1]
    plot_series(X[0, :, 0], num_predictions = Y_pred.shape[1])
    plt.plot(np.arange(n_steps, n_steps + ahead), Y[0, :, 0], "ro-", 
             label="Actual")
    plt.plot(np.arange(n_steps, n_steps + ahead), Y_pred[0, :, 0], "bx-", 
             label="Forecast", markersize=10)
    plt.axis([0, n_steps + ahead, -1, 1])
    plt.legend(fontsize=14)
    return
```

Make training and validation data. Test data will be generated and tested later. Feeding 50 data points (X_train) to predict the next 10 (Y_train).

```python
np.random.seed(0) # not 42, as it would give the first series in the train set
n_steps = 50
series = simulate_time_series(10000, n_steps + 10)
X_train, Y_train = series[:7000, :n_steps], series[:7000, -10:, 0]
X_valid, Y_valid = series[7000:9000, :n_steps], series[7000:9000, -10:, 0]
plt.figure()
plt.plot(X_train[1])
plt.show()
```

Create Deep learning forecasting model

```python
np.random.seed(42)
tf.random.set_seed(42)

model = Sequential()
model.add(SimpleRNN(20, return_sequences=True, input_shape=[None, 1]))
model.add(SimpleRNN(20))
model.add(Dense(10))
print(model.summary())
plot_model(model)
```

Model Training

```python
model.compile(loss="mse", optimizer="adam")
model_training = model.fit(X_train, Y_train, epochs=20,
                    validation_data=(X_valid, Y_valid))
```

Model Evaluation - model is learning!

```python
plt.figure(figsize=(10, 5))
plt.plot(model_training.history['loss'])
plt.plot(model_training.history['val_loss'])
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train Data', 'Validation Data'])
plt.show()
```

Feeding 50 samples and receiving 10 predictions.

```python
series = simulate_time_series(10, 50 + 10) 
X_test, Y_test = series[:, :50, :], series[:, -10:, :]
Y_pred = model.predict(X_test)[..., np.newaxis]
```

Plot forecasted predictions from test data.

```python
plt.figure(figsize = (15, 25))
for i in range(5):
    plt.subplot(5, 1, i + 1)
    plot_multiple_forecasts(X_test[i].reshape([1, 50, 1]), 
                            Y_test[i].reshape([1, 10, 1]), 
                            Y_pred[i].reshape([1, 10, 1]))
    plt.title('Forecast Time Series')
plt.show()
```
